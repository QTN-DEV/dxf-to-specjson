```python
folders = [
    "dress_150/",
    "dress_sleeveless_2550/",
    "jacket_2200/",
    "jacket_hood_2700/",
    "jacket_hood_sleeveless_150/",
    "jacket_sleeveless_150/",
    "jumpsuit_150/",
    "jumpsuit_sleeveless_2000/",
    "pants_straight_sides_1000/",
    "skirt_2_panels_1200/",
    "skirt_4_panels_1600/",
    "skirt_8_panels_1000/",
    "skirt_waistband_150/",
    "tee_2300/",
    "tee_hood_150/",
    "tee_sleeveless_1800/",
    "wb_dress_sleeveless_2600/",
    "wb_jumpsuit_sleeveless_150/",
    "wb_pants_straight_1500/"
]
import os
import json
import pandas as pd


main_dir = ''  


main_folders = folders  


data = []

for folder in main_folders:
    folder_path = os.path.join(".", folder)

    if not os.path.isdir(folder_path):
        print(f"Folder utama tidak ditemukan: {folder_path}")
        continue
    

    subfolders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]
    
    for subfolder in subfolders:
        subfolder_path = os.path.join(folder_path, subfolder)
        json_file_path = os.path.join(subfolder_path, 'specification.json')
        

        if not os.path.isfile(json_file_path):
            print(f"File specification.json tidak ditemukan di: {subfolder_path}")
            continue

        try:
            with open(json_file_path, 'r') as f:
                spec = json.load(f)
        except json.JSONDecodeError as e:
            print(f"Error membaca JSON di {json_file_path}: {e}")
            continue
        
        # Navigasi ke 'pattern' -> 'panels'
        try:
            panels = spec['pattern']['panels']
        except KeyError as e:
            print(f"Key {e} tidak ditemukan di {json_file_path}")
            continue
        
        # Iterasi melalui setiap panel
        for sewing_name, attributes in panels.items():
            try:
                translation = attributes['translation']
                rotation = attributes['rotation']
                vertices = attributes['vertices']
                edges = attributes['edges']
                temporary_vertices = []
                temporary_curves = []
                for i in range(len(edges)):
                    temporary_vertices.append(vertices[edges[i]["endpoints"][0]])
                    temporary_curves.append(edges[i].get("curvature", (None, None)))
                for i in range(10-len(temporary_vertices)):
                    temporary_vertices.append((None,None))
                    temporary_curves.append((None,None))
                # Pastikan translation dan rotation memiliki setidaknya 3 item
                if len(translation) < 3 or len(rotation) < 3:
                    print(f"Data translation atau rotation kurang di {json_file_path} untuk panel {sewing_name}")
                    continue
                
                # Buat dictionary untuk baris ini
                row = {
                    'name_garment': subfolder,
                    'sewing_name': sewing_name,
                    't0': translation[0],
                    't1': translation[1],
                    't2': translation[2],
                    'r0': rotation[0],
                    'r1': rotation[1],
                    'r2': rotation[2],
                    # 'vertices_length': len(vertices),
                    # 'edges_length': len(edges)
                }
                for i in range(10):
                    row[f'vertices_x{i}'] = temporary_vertices[i][0]
                    row[f'vertices_y{i}'] = temporary_vertices[i][1]
                    row[f'curves_x{i}'] = temporary_curves[i][0]
                    row[f'curves_y{i}'] = temporary_curves[i][1]
                data.append(row)
            
            except KeyError as e:
                print(f"Key {e} tidak ditemukan di panel {sewing_name} di {json_file_path}")
                continue

# 5. Buat DataFrame dari data yang terkumpul
df = pd.DataFrame(data)

# (Opsional) Tampilkan DataFrame
print(df.head())

# (Opsional) Simpan DataFrame ke file CSV
output_csv = 'panels_data.csv'
df.to_csv(output_csv, index=False)
print(f"Data berhasil disimpan ke {output_csv}")

```

```python
import json

def extract_failed_files(json_data):
    """
    Extracts the union of files listed under 'fails' in the JSON structure
    from 'crashes' to 'static_equilibrium'.

    Parameters:
        json_data (dict): Parsed JSON data.

    Returns:
        set: A set containing the union of all failed files.
    """
    failed_files = set()

    # Navigate to the 'fails' section
    fails_section = json_data.get('sim', {}).get('stats', {}).get('fails', {})
    
    # Iterate through the relevant keys
    for key in ['crashes', 'intersect_colliders', 'intersect_self', 'static_equilibrium']:
        if key in fails_section:
            failed_files.update(fails_section[key])
    
    return failed_files

# # Example usage
# if __name__ == "__main__":
#     # Assuming the JSON structure is saved in a file named 'data.json'
#     with open('data.json', 'r') as file:
#         json_data = json.load(file)
    
#     failed_files = extract_failed_files(json_data)
#     print("Union of failed files:", failed_files)
```

```python
# exclude left_1, left_2 () sama dengan left_0; right_1, right_2 karena redundant
# exclude lbsleeve karena sama dengan rfsleeve,  rbsleeve karena sama dengan lfsleeve
# exclude skirt_back karena udah ada skirt_front
# Lback karena ada Rfront, Rback karena ada Lfront
# wb_back karena ada wb_front
excluded_label = ["left_1","left_2","right_1","right_2","lbsleeve","rbsleeve","skirt_back","Lback","Rback","wb_back"]
# Membuat DataFrame baru yang tidak mengandung name_garment dari list lstlst
df_clean = df_filtered[~df_filtered['sewing_name'].isin(excluded_label)]
df_clean = df_clean.reset_index(drop=True)
output_csv = '../panels_data_clean.csv'
df_clean.to_csv(output_csv, index=False)
df_clean
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Misalkan df_filtered adalah DataFrame yang sudah ada
# 'sewing_name' adalah kolom label

# Langkah 1: Split data menjadi train (80%) dan temp (20%)
train_df, temp_df = train_test_split(
    df_clean,
    test_size=0.2,  # 20% untuk temp
    stratify=df_clean['sewing_name'],  # Stratifikasi berdasarkan 'sewing_name'
    random_state=2024  # Untuk reproduktibilitas
)

# Langkah 2: Split temp menjadi dev (10%) dan test (10%)
# Karena temp adalah 20% dari data awal, kita bagi temp menjadi 50% dev dan 50% test
dev_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,  # 50% dari temp untuk test, 50% untuk dev
    stratify=temp_df['sewing_name'],  # Stratifikasi berdasarkan 'sewing_name'
    random_state=2024  # Untuk reproduktibilitas
)

# Opsional: Reset index jika diperlukan
train_df = train_df.reset_index(drop=True)
dev_df = dev_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# Menampilkan ukuran masing-masing set
print(f"Train set: {train_df.shape[0]} sampel")
print(f"Dev set: {dev_df.shape[0]} sampel")
print(f"Test set: {test_df.shape[0]} sampel")

# Opsional: Memeriksa distribusi 'sewing_name' di setiap set
print("\nDistribusi 'sewing_name' di Train set:")
print(train_df['sewing_name'].value_counts(normalize=True))

print("\nDistribusi 'sewing_name' di Dev set:")
print(dev_df['sewing_name'].value_counts(normalize=True))

print("\nDistribusi 'sewing_name' di Test set:")
print(test_df['sewing_name'].value_counts(normalize=True))
```
